```python
import pandas as pd

# Create dataset
data = {
    "OrderID": [1001,1002,1003,1004,1005,1006,1007,1008,1009,1010],
    "Date": ["2024-01-05","2024-01-06","2024-01-07","2024-02-01","2024-02-03",
             "2024-02-10","2024-03-02","2024-03-05","2024-03-10","2024-03-15"],
    "Category": ["Clothing","Clothing","Shoes","Clothing","Shoes","Accessories",
                 "Clothing","Accessories","Shoes","Clothing"],
    "Product": ["T-Shirt","Jeans","Sneakers","Jacket","Boots","Bag",
                "T-Shirt","Belt","Sneakers","Jeans"],
    "Sales": [500,1200,2000,2500,1800,900,700,400,2200,1500],
    "Quantity": [5,3,4,2,3,6,7,5,2,4],
    "City": ["Stockholm","Stockholm","Gothenburg","Malmo","Stockholm",
             "Gothenburg","Stockholm","Malmo","Stockholm","Gothenburg"]
}

df = pd.DataFrame(data)

df.head()
```

```python
df["Sales"].sum()
```
