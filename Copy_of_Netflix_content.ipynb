```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Load the data set

```python
df = pd.read_csv("netflix_full.csv")
df.head()
```

Understanding the data

```python
df.info()
```

```python
df.isnull().sum()
```

Clean the data

```python
df['country'] = df['country'].fillna('Unknown')
```

```python
df = df.drop(columns=['creator'])
```

```python
df.isnull().sum()
```

```python
df['genres']= df['genres'].fillna('Unknown')
```

```python
df.isnull().sum()
```

Movies vs TV Shows

```python
type_counts = df['type'].value_counts()
print(type_counts)
```

```python
type_counts.plot(kind='bar')
plt.title("Movies vs TV Shows")
plt.show()
```

Content over time

```python
year_counts = df['year'].value_counts().sort_index()
print(year_counts)
```

```python
year_counts.plot(kind='line')
plt.title("Content Released Over Time")
plt.show()
```

Top countries

```python
country_counts = df['country'].value_counts().head(10)
print(country_counts)
```

```python
country_counts = df[df['country'] != 'Unknown']['country'].value_counts().head(10)
print(country_counts)
```

```python
country_counts.plot(kind='bar')
plt.title("Top 10 Countries")
plt.xticks(rotation=45)
plt.show()
```

Ratings analysis

```python
rating_counts = df['rating'].value_counts()
print(rating_counts)
```

```python
rating_counts.plot(kind='bar')
plt.title("Content Ratings")
plt.xticks(rotation=45)
plt.show()
```

Genres


```python
genre_counts = df['genres'].value_counts().head(10)
print(genre_counts)
```

```python
genre_counts.plot(kind='bar')
plt.title("Top Genres")
plt.xticks(rotation=90)
plt.show()
```

Key insights

*  Netflix has slightly more Movies than TV Shows, but the difference is very small.
*   Netflix content shows a strong growth trend from 2012 to 2022, followed by a decline in 2023.
*   Netflix content is heavily dominated by the United States, but there is also strong international contribution, especially from Asia.
*   Netflix content is heavily focused on teen and adult audiences, especially TV-14 and TV-MA ratings, which together dominate the platform.
*   Netflix content is heavily dominated by comedy and romance-related genres, especially Stand-Up Comedy and Romantic content.
